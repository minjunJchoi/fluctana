#### Step #0
Import modules.

```python
%matplotlib inline 
import sys, os
sys.path.append(os.pardir)
from fluctana import *
```

#### Step #1 
Make an instance for calculation and add data set.

```python
A = FluctAna()
A.add_data(KstarEcei(shot=18597, clist=['ECEI_G0101-0408']), trange=[1.78, 1.81])
#A.add_data(KstarMds(shot=18597, clist=['MC1T06']), trange=[1.78, 1.81], norm=0, res=0.000002)
```

You can check data set and channels as follows

```python
A.list_data()
```

A.add_data() takes an instance of 'KstarEcei' class as an input, and all the data and information is saved in that instance. It can be accessed as follows (A.Dlist[data set number])

```python
A.Dlist[0]
```

Data is saved in 'data' of A.Dlist[0]. It has (channel number X data points) dimension

```python
print  A.Dlist[0].data.shape
```

The first channel data is (note that the data is normalized (default))

```python
print A.Dlist[0].data[0,:]
```

Time base is

```python
print A.Dlist[0].time
```

Channel positions are (in channel number ordering)

```python
print A.Dlist[0].rpos # R [m]
print A.Dlist[0].zpos # z [m]
```

#### Step #2
Plot a few channels (channel number 0, 10, 20) in data set number 0 in multiple windows.

```python
A.mplot(dnum=0,cnl=[0,10,20],type='time')
```

Plot a few channels (channel number 0, 10, 20) in data set number 0 in single window (overlapping)

```python
A.oplot(dnum=0,cnl=[0,10,20],type='time')
```

Plot spectrums of a few channels (channel number 0, 10, 20) in data set number 0 in multiple windows

```python
A.spec(dnum=0,cnl=[0,10,20],climits=[-100,-30],flimits=[0,50000])
```

Plot (interactively) ECE image of data set number 0 with sample channel number 0

```python
A.iplot(dnum=0,snum=0,vlimits=[-0.1,0.1])
```
